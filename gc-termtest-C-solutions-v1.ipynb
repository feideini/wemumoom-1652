##### Question 5

Find the length of the following curve.
$$
y=\int_{0}^{x}\sqrt{\sec^{4}t-1}\,dt,-\frac{\pi}{3}\leq{}x\leq\frac{\pi}{3}
$$
Remember that according to the Fundamental Theorem of Calculus, if $g(x)=\int_{a}^{x}f(t)\,dt$, then $g'(x)=f(x)$.

$$L=\int_{-\frac{\pi}{3}}^{\frac{\pi}{3}}\sqrt{1+\left(\sqrt{\sec^{4}x-1}\right)^{2}}\,dx=\int_{-\frac{\pi}{3}}^{\frac{\pi}{3}}\sec^{2}x\,dx=\left[\tan{}x\right]_{-\frac{\pi}{3}}^{\frac{\pi}{3}}=2\sqrt{3}$$
